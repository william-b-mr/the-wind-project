```python
import requests
import pandas as pd
import numpy as np
import sys
import urllib.parse
import matplotlib.pyplot as plt
```

```python
key = "5bb3639a-ef5a-11ed-8b7f-0242ac130002-5bb363fe-ef5a-11ed-8b7f-0242ac130002"
BASE_URL = 'https://api.stormglass.io/v2/'

#latitude and longitude of Montijo
lat = 38.687616
lon = -8.855195
```

```python
def get_soil_moisture(lat,lon):
    """
    Returns a df with the forecast of soil moisture at the surface, 10cm, 40cm and 100cm of depth
    """
    
    url = urllib.parse.urljoin(BASE_URL,'bio/point')

    response = requests.get(
      url,
      params={
        'lat': lat,
        'lng': lon,
        'params': ','.join(['soilMoisture', 'soilMoisture10cm', 'soilMoisture40cm', 'soilMoisture100cm']),
      },
      headers={
        'Authorization': key
      }
    )
    
    #Getting a pandas df for the response
    #data_df = pd.DataFrame(response.json()['hours'])
    
    
    return response.json()
```

```python
# Make an API call for Montijo's lat and lon
moisture_json = get_soil_moisture(lat,lon)

# Make a df out of it
moisture_df = pd.DataFrame(moisture_json['hours'])
```

```python
moisture_df
```

As with the wind data, the entries of each parameter are dictionaries that contain the value predicted at different weather institutes. Therefore we'll apply some changes, by transforming these into columns with the respective values e.g. noaaSurface sgSurface will show the correspondent values from the soilMoisture column

```python
moisture_df[['noaaSurface','sgSurface']] = pd.json_normalize(moisture_df['soilMoisture'])
```

```python
moisture_df
```

Now applying these to the remaining columns:

```python
moisture_df[['noaa10cm','sgSurface10cm']] = pd.json_normalize(moisture_df['soilMoisture10cm'])
moisture_df[['noaa40cm','sgSurface40cm']] = pd.json_normalize(moisture_df['soilMoisture40cm'])
moisture_df[['noaa100cm','sgSurface100cm']] = pd.json_normalize(moisture_df['soilMoisture100cm'])
```

```python
moisture_df_clean = moisture_df.drop(columns = ['soilMoisture','soilMoisture100cm','soilMoisture10cm','soilMoisture40cm'])
```
